### Definiciones de rango

1.  La dimensionalidad del espacio de columnas.
2.  La cantidad de columnas que forma un set que es linealmente independiente.
3.  El rango de una matriz es el mismo para sus filas y sus columnas.
El rango negativo no existe, una matriz no puede tener varios rangos, el rango no tiene un maximo puede irse hasta infinito, pero esta acotado por el numero mas chico ya sea de filas o columnas. El rango se expresa como r(R).

Si conoces el rango de dos matrices  A, B.
Puedo conocer el rango de A + B ?
El rango de una suma puede ser mayor que los rangos de las matrices individuales.

El rango de A x B?
El rango de una multiplicacion no puede ser mayor que el rango mas grande de las matrices multiplicadas.
