### Para que sirven las matrices en algebra lineal ?

Su funcion principal del algebra lineal es resolver sistemas de equaciones, 
armar regresiones lineales, 
calcular autovalores y autovectores, 
y resolver problemas de optimizacion.

Aun con computadore se usa el algebra lineal para resolver sistemas de ecuaciones.

![Sistema de ecuaciones](image13.png)

En el cual tenemos tres variables x, y, z. El sistema son equaciones lineales. Nada que no sea lineal.

Se pueden ordenar como matrices, pero primero se deben ordenar de la siguiente formula.
![Sistema de ecuaciones matricial](image14.png)

Acomodamos todas la ecuaciones para que queden alineadas todas las variables.

Agregamos los valores que no existen para las variables que faltan.
![Sistema de ecuaciones relleno con ceros](image15.png)

Ordenamos los coeficientes de las variables en una matriz (en este caso "A").
![Matriz A](image16.png)

Ordenamos las variables en un vector (en este caso 'X') 
Tomamos las constantes y las metemos en un vector (en este caso "v").
![Vector v](image17.png)

Debemos encontrar valores para que la matriz 'X' que llegue a un vector 'v'.
![Sistema de ecuaciones](image18.png)

Lo que hacemos con el sistema de ecuaciones es resolverlo:
Es decir buscamos un vector 'X' que sufra una transformacion 'A' y nos retorne el vector resultante 'V'.

Encontrar formas que nos faciliten hacer el calculo y permitan encontrar los resultados.
